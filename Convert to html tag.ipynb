```python
def convertRowToHtml(id, content):
    arr = ["<tr>", "<td class='sign_id' >", id, 
           "</td>", "<td> <img src='data:image/png;base64,  ", 
           content, "'/></td>", "<td class='is_omni' >", 
           "<input type='checkbox' >", "</td>", "</tr>"]
    return " ".join(arr)
```
